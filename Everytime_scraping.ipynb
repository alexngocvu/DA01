```python
from google_play_scraper import Sort, reviews

# Set the app ID for Everytime app
app_id = 'com.everytime.v2'

# Fetch the reviews
review_data, _ = reviews(
    app_id,
    lang='en',  # Language of reviews (English)
    country='us',  # Country (United States)
    sort=Sort.NEWEST,  # Sort reviews by newest
    count=100  # Number of reviews to fetch; adjust as needed
)

# Display the review text
for i, review in enumerate(review_data):
    print(f"Review {i + 1}: {review['content']}")
    print(f"Rating: {review['score']}")
    print(f"Date: {review['at']}")
    print("-" * 80)
```

```python
from google_play_scraper import Sort, reviews
import csv

# Function to save reviews to a CSV file with UTF-8 BOM
def save_reviews_to_csv(app_id, lang, filename):
    # Fetch the reviews
    review_data, _ = reviews(
        app_id,
        lang=lang,  # Language (set to 'en' for English or 'ko' for Korean)
        country='us',  # Country (United States)
        sort=Sort.NEWEST,  # Sort reviews by newest
        count=100  # Number of reviews to fetch
    )

    # Write to CSV file with UTF-8 BOM encoding
    with open(filename, mode='w', newline='', encoding='utf-8-sig') as file:
        writer = csv.writer(file)
        # Write the header
        writer.writerow(['Review', 'Rating', 'Date'])

        # Write the review data
        for review in review_data:
            writer.writerow([review['content'], review['score'], review['at']])

    print(f"Reviews saved to {filename}")

# App ID for Everytime app
app_id = 'com.everytime.v2'

# Save English reviews
save_reviews_to_csv(app_id, lang='en', filename='everytime_reviews_en.csv')

# Save Korean reviews
save_reviews_to_csv(app_id, lang='ko', filename='everytime_reviews_ko.csv')
```
